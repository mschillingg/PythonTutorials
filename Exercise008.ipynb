### Import libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read data

```python
mdff = pd.read_csv("datasets/008ExerciseFile.csv")
```

### Display data

```python
display(mdff)
```

### Plotting x against y

```python
mdff.plot(x='rel_time', y='m_level', kind='scatter', color='r')
```

```python
sns.set_theme() # set theme to seaborn's default
```

```python
sns.relplot(data=mdff, x='rel_time', y='m_level')
```

### First level polynomial

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level')
```

The bow tie shape of the error bounds of this line above shows us that the slope isn't as accurate as it could be.

```python
sns.residplot(data=mdff, x='rel_time', y='m_level')
```

There is a pattern to these residuals, so there's something going on in the data that isn't being accurately represented by a first order polynomial. Let's try a second order.

### Second order

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level', order=2)
```

The error bounds of this line are still pretty wide.

```python
sns.residplot(data=mdff, x='rel_time', y='m_level', order=2)
```

There is still a pattern the second order polynomial's residuals. Let's try a third order.

### Third level

```python
sns.lmplot(data=mdff, x='rel_time', y='m_level', order=3)
```

```python
sns.residplot(data=mdff, x='rel_time', y='m_level', order=3)
```

No more pattern to the residuals! The points being randomly distributed around 0 means that this is our best fit polynomial.

### Pretty plot

```python
sns.set(style="whitegrid")
sns.lmplot(data=mdff, x='rel_time', y='m_level', order=3, line_kws={'color':'red'})
plt.ylabel("Relative Time")
plt.xlabel("Melatonin Level")
plt.title("Melatonin Level Over Time")
```
