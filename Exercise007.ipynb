### Import libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read .csv file

```python
myDataFromFile = pd.read_csv("datasets/007ExerciseFile.csv")
```

```python
display(myDataFromFile)
```

### Histogram

```python
sns.displot(myDataFromFile, x="Mutation_rate", hue="Strain", kind="hist", bins=50, alpha=0.3)
```

### KDE plot

```python
sns.displot(myDataFromFile, x="Mutation_rate", hue="Strain", kind="kde")
```

```python
sns.displot(myDataFromFile, x="Mutation_rate", hue="Strain", kind="kde", fill=True, alpha=0.2)
```

### Violin plot

```python
sns.violinplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
```

### Boxplot

```python
sns.boxplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
```

### Boxplot with overlaid strip plot

```python
sns.boxplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
sns.stripplot(data=myDataFromFile, x="Mutation_rate", y="Strain", alpha = 0.4)
```

## Dr. Fauci, I present to you the Boxplot with Overlaid Strip Plot:

```python
sns.boxplot(data=myDataFromFile, x="Mutation_rate", y="Strain")
sns.stripplot(data=myDataFromFile, x="Mutation_rate", y="Strain", alpha = 0.3)
plt.xlabel("Mutation Rate")
y=[0,1,2,3]
labels=["Delta","Gamma","Beta","Alpha"]
plt.yticks(y, labels)
```

### This plot gives a nice visualization of the dataâ€“the distance between the median lines on each of the boxes highlights the significant differences between the mutation rates of each of the strains.
